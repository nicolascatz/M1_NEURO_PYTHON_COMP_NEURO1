# Matplotlib Library

This library has functions for visualization, like line plots, scatter plots, etc. See https://matplotlib.org/stable/users/explain/quick_start.html for the first steps. Other tutorials can be found there: https://matplotlib.org/stable/tutorials/index

Here we consider fMRI/BOLD data as time series.

```python
import numpy as np
import matplotlib.pyplot as plt

# time vector
t = np.arange(10)
# data vector
x = np.load('BOLD_movie.npy')[20,20,10,:]

plt.plot(t, x)
plt.savefig('plot_x')
```

We can add labels, change the ofnt size, etc.

```python
# create a new figure
plt.figure()
# plot values of x as a function of those of t
plt.plot(t, x, label='x')
# uses the label keyword in plot functions
plt.legend()
# axe labels
plt.xlabel('time')
plt.ylabel('data')
# compact formatting
plt.tight_layout()
```

```python
# another vector
y = np.load('BOLD_movie.npy')[15,20,10,:]

# change the default font size to 16
import matplotlib as mpl
mpl.rcParams['font.size'] = 16

plt.figure(figsize=[12,5])
plt.subplot(1,2,1)
plt.plot(t, x, color='red', label='x')
plt.plot(t, y, color='blue', label='y')
plt.legend(fontsize=12)
plt.xlabel('time')
plt.ylabel('data')
plt.subplot(1,2,2)
plt.plot(x, y, ls='', marker='x', color='magenta')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
plt.savefig('ex_subplot')
```

The function plot can also display the markers without the line by using arguments (see the documentation https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

# Plotting a function

Now considering a function $f(t)$ of a time variable $t$, we can plot the function to then see its profile, extrema (minima, maxima), etc.

```python
# definition of the function
def f(t):
    # return calculation
    return t / (1 + t**2)
```

```python
# definition of vector of time for x-axis: 20 points uniformaly spaced from -10.0 to 10.0
vt = np.linspace(-10.0, 10.0, 20)

print(vt)
```

Instead of passing a single argument to the function, we can pass the vector $vt$ directly and the function output is the images of the corresponding arguments in $vt$.

```python
print(f(vt))
```

We can now plot $f(t)$ as a function of $t$ for each t in $vt$.

```python
plt.plot(vt, f(vt))
plt.xlabel('$t$')
plt.ylabel('function $f(t)$')
plt.show()
```

The function doesn't seem smooth, we don't have a sufficient resolution for the points in $vt$, let's use 100 points.

```python
vt = np.linspace(-10.0, 10.0, 100)

plt.plot(vt, f(vt))
plt.xlabel('$t$')
plt.ylabel('function $f(t)$')
plt.savefig('ex_func')
```

Finally, let's find the maximum of the plotted curve (in the range given by $vt$):

```python
# index of largest element of f(vt)
i_t = np.argmax(f(vt))

print(i_t, vt[i_t])
```

```python
plt.plot(vt, f(vt), label='curve')
plt.plot(vt[i_t], f(vt[i_t]), 'xr', ms=10, label='max') # xr stands for red cross, ms for marker size
plt.legend(fontsize=12) # to show labels associated with plotted elements
plt.xlabel('$t$')
plt.ylabel('function $f(t)$')
plt.show()
```

# Plotting a distribution

Several options are available to represent distributions:
- `histogram`, see https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html
- `boxplot`, see https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html
- `violinplot`, see https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html

```python
# load distribution of BOLD values in a 3D image
x = np.load('BOLD_image.npy')
print(x.shape)
```

The keyword `bins` is the number of bars in the histogram. Check https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html for details on other options.

```python
plt.figure()
plt.hist(x.flatten(), bins=40)
plt.xlabel('value')
plt.ylabel('count')
plt.tight_layout()
plt.savefig('ex_hist')
```

By default, the median is indiciated in color, the box represents the extension between the 1st and 3rd quartiles, lines indicate the limit 

```python
plt.figure()
plt.boxplot(x.flatten())
plt.xlabel('value')
plt.ylabel('count')
plt.tight_layout()
plt.savefig('ex_box')
```

```python
plt.figure()
plt.violinplot(x.flatten())
plt.xlabel('value')
plt.ylabel('count')
plt.tight_layout()
plt.savefig('ex_violin')
```

# Plotting an image

Here we use `imshow`, check https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html for details.

```python
# load distribution of BOLD values in a 3D image
img = np.load('BOLD_image.npy')
print('shape:', img.shape)
# get the number of voxels
ni, nj, nk = img.shape

plt.figure()
# plot the image
plt.imshow(img[25,:,:])
# add a colorbar to link the color plotted to values
plt.colorbar()
# add labels to axes
plt.ylabel('j')
plt.xlabel('k')
plt.show()
```

In the plot above, we want to put y on the horizontal axis and z on the vertical axis. We can transpose the image (as a matrix) and choose the origin $(0,0)$ at the bottom left.

```python
plt.figure()
# plot the image
plt.imshow(img[25,:,:].T, origin='lower')
# add a colorbar to link the color plotted to values
plt.colorbar()
# add labels to axes
plt.xlabel('j')
plt.ylabel('k')
plt.show()
```

To go further, we can change the tick labels for the metric space instead of voxel indices, change the color map.

```python
plt.figure()
# plot the image
plt.imshow(img[25,:,:].T, origin='lower', cmap='hot')
# add a colorbar to link the color plotted to values
plt.colorbar()
# add labels and ticks to axes
x_tick_min, x_tick_max, y_tick_min, y_tick_max, z_tick_min, z_tick_max = np.load('BOLD_coord.npy')
vx = np.linspace(x_tick_min, x_tick_max, ni, dtype=int)
vy = np.linspace(y_tick_min, y_tick_max, nj, dtype=int)
vz = np.linspace(z_tick_min, z_tick_max, nk, dtype=int)
plt.xticks(range(0,nj,17), vy[range(0,nj,17)])
plt.yticks(range(0,nk,20), vz[range(0,nk,20)])
plt.xlabel('y')
plt.ylabel('z')
plt.show()
```
