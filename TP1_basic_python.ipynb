The object `range` provides a way to get indices from an integer to another integer. If only one integer $n>0$ is specified, the range is between $0$ and $n-1$. When two integers $0 < n_1 < n_2$ are provided, the range is between $n_1$ included and $n_2$ excluded. A third argument can be provided to specify the step, which is $1$ by default.

````python
for i in range(4):
    print(i)
````

````python
for i in range(1, 8):
    print(i)
````

````python
for i in range(1, -4, -1):
    print(i)
````

````python
for i in range(len(l)):
    print(l[i])
````

## Numpy Library

To use functions in a library, we import it and call its member functions as follows. Here we create an "alias" or short name for numpy because it is shorter to write.

````python
import numpy as np

x = np.array([0, 2, 8, -3])

print(x)
````

````python
x.dtype
````

The attribute `dtype` indicates the type of elements in the array. All items must be of the same type.

````python
print(np.array([0, 2, 8.0, -3]))
print(np.array([0, 2, 8.0, -3]).dtype)
````

The advantage of arrays over list is that mathematical operations can be used on them (they have been made for that...).

````python
x = np.array([0, 2, 8, -3])
y = np.array([10, 4, 6, 3])

print(x)
print(y)
print(x + y)
````

````python
print(x * y)
````

The operator `+` has a different behavior for arrays, lists and strings.

````python
print(x.tolist() + y.tolist())
````

````python
print('comp' + 'neuro')
````

Arrays can be constructed with arbitrary dimensions ('shape'), some operations require that they have matching shapes, see above or the scalar product below. Recall that the dot product above is the sum of the product of corresponding items:
$$ \sum_i x_i y_i $$

````python
print(np.dot(x, y))
print(x @ y)
print(np.sum(x*y))
````

````python
print(x.shape, y.shape)
````

````python
z = np.array([2, 4])

print(np.dot(x, z))
````

Note that a lot of operators (like `*` for multiplication, `**` for power) that also work for arrays (search for 'linalg' in the doc), like matrix operations.

````python
print('square of 3 =', 3**2, '; 4 to the power of 1.5 =', 4**1.5, '; x squarred:', x**2)
````

Let's now construct a vector of number using a generator function for evenly spaced number. We can access the elements using their correponding indices (or ranges of indices called 'slices'). The arguments `:` means that we want all dimensions for the corresponding index. It can be preceeded by a number to indicate a range (note that the last index is excluded, and the first index is zero). Negative indices start from the end.

````python
# array with values determined by a step, like from 0.0 included to 2.0 excluded with step 0.4
v = np.arange(0.0, 2.0, 0.4)
print(v)
````

````python
print(v[0])
print(v[1:3])
print(v[-1:])
````

Note that `numpy.arange` can also be used as `range` for get range of indices, but it is an array object (`numpy.ndarray`), not a `range` object.

````python
A = np.arange(6)
print(A)
print(type(A))

print(type(range(6)))
````

Arrays can be reshaped; note that several orders are possible (like rolling over early versus late indices first), but don't use non-default options unless you know what you are doing! 

````python
B = A.reshape([2,3])
print(B)
````

````python
print(B.flatten())
````

The target shape can involve `-1` for one index, which is then calculted from the total number of elements and the specified desired dimensions.

````python
print(A.reshape([2,-1]))
````

You can save an array into a file on your drive using the `numpy.save` function, the default file extension is 'npy'. You can then get it back using `numpy.load`.

````python
np.save('A.npy', A)
````

````python
C = np.load('A.npy')

# did we recover the same data?
print(A==C)
````

But the target shape must be appropriate, i.e. matching the total number of elements in the array. Otherwise an error is raised.

````python
A.reshape([4,-1])
````

### 2D array

2-dimensional arrays are the same as extensions of the above examples with 2 indices. Here we generate a random array using the numpy library. The transpose `T̀ means to swap the two indices.

````python
D = np.array([[0, 1, 2], [3, 4, 5]])

print('shape of D:', D.shape)
print(D)
````

````python
i = 1
j = 2
print(f'value of D[i,j] for i={i}, j={j}:'.format(i,j), D[1, 2])
````

Another way to access elements is to provide boolean arrays of matching shapes, such that only indices with `True` will be retained.

````python
sel_j = np.arange(74)<10
print(sel_j)
````

````python
M[:,sel_j,:].shape
````

````python
# select matrix elements
sel_elem = np.tri(74,62,0,dtype=bool)
print(sel_elem)
print(sel_elem.shape)
print('number of True in boolean array:', sel_elem.sum())
````

````python
M[:,sel_elem].shape
````

Yet another way to access elements is to provide a list of indices.

````python
sel_elem2 = np.tril_indices(74,0,62)
print(sel_elem2)
print(type(sel_elem2))
````

````python
M[:,sel_elem2[0],sel_elem2[1]].shape
````

Do we get the same elements using these two methods?

````python
np.all(M[:,sel_elem]==M[:,sel_elem2[0],sel_elem2[1]])
````

CAREFUL! the following behavior is a bit tricky, when extracting elements using boolean arrays... So prefer arrays or lists of indices in general.

````python
M2 = M[0,:10,:10] # keep only 10 dimensions for all indices

sel_ind = np.arange(10)<3

print(M2[sel_ind,sel_ind], 'of shape', M2[sel_ind,sel_ind].shape)
print(M2[sel_ind,:][:,sel_ind], 'of shape', M2[sel_ind,:][:,sel_ind].shape)
````
